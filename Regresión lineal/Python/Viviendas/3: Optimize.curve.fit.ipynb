# Regresión lineal múltiple con el método curve_fit() de Scipy

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import time
import sys
```

```python
datos = pd.read_csv("./kc_house_data.csv")
    
datos.head()
```

```python
datos=datos.drop(['id','date','sqft_basement'],axis=1)
```

```python
#Variables independientes
X = datos.drop(['price'], axis=1)
X = pd.get_dummies(X)
X.head()
```

```python
#Variables dependientes
Y = datos['price']
Y.head()
```

```python
t1=time.time()
```

Puesto que queremos hacer un análisis de regresión multiple, es decir, con varias variables independientes, tenemos que pasarle como parámetro a la función curve_fit una función func en la cual se establece la relación entre las variables independientes.

```python
x1 = X.loc[:,'bedrooms']
x2 = X.loc[:,'bathrooms']
x3 = X.loc[:,'sqft_living']
x4 = X.loc[:,'sqft_lot']
x5 = X.loc[:,'floors']
x6 = X.loc[:,'waterfront']
x7 = X.loc[:,'view']
x8 =  X.loc[:,'condition']
x9 = X.loc[:,'grade']
x10 = X.loc[:,'sqft_above']
x11 = X.loc[:,'yr_built']
x12 = X.loc[:,'yr_renovated']
x13 = X.loc[:,'zipcode']
x14 = X.loc[:,'lat']
x15 = X.loc[:,'long']
x16 = X.loc[:,'sqft_living15']
x17 = X.loc[:,'sqft_lot15']
```

```python
def func(X,a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13,a14,a15,a16,a17):
 
    return a0+x1*a1+x2*a2+x3*a3+x4*a4+x5*a5+x6*a6+x7*a7+x8*a8+x9*a9+x10*a10+x11*a11+x12*a12+x13*a13+x14*a14+x15*a15+x16*a16+x17*a17
```

Ahora ya tenemos todo lo necesario para realizar la llamada al método

```python
resultado= curve_fit(f= func, xdata = X,ydata= Y)
```

```python
#Estudio tiempo de ejecucion

t2=time.time()
tiempo = float(t2-t1)

print('Regresion lineal utilizando Optimize.curve.fit()')
print("Tiempo de ejecucion: {} segundos".format(tiempo))
```
